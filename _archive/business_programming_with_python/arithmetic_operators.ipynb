# Arithmetic Operators

In Python, operators are special symbols that designate that some sort of 
computation should be performed. The values that an operator acts on are called 
operands. The following table list the arithmetic operators

| Operation | Description           | Example       |  Result  |
|:---------:|-----------------------|---------------|----------|
| ```+```   | Addition              | ```3 + 2.2``` | ```5.2```|
| ```-```   | Subtraction           | ```5 - 2```   | ```3```  |
| ```*```   | Multiplication        | ```3 * 8```   | ```24``` |
| ```**```  | Raise to the power of | ```3 ** 2```  | ```9```  |
| ```/```   | Division              | ```5/2```     | ```2.5```|
| ```//```  | Integer Division      | ```5/2```     | ```2```  |
| ```%```   | Modulo                | ```5%2```     | ```1```  |

Regular arithmetic rules apply! This includes the [order of execution](https://www.khanacademy.org/math/pre-algebra/pre-algebra-arith-prop#pre-algebra-order-of-operations)! But just like in math, you can use parentheses to decide what happens when. Let's see how you can accomplish this in Python:



````python
3 + 2.2 * 2
````

````python
(3 + 2.2) * 2
````

```3 + 2.2 * 2``` is called an expression.  When you evaluate an expression you can store the result into a variable

````python
result = 3 + 2.2 * 2
print(result)
````

Can perform arithmetic using variables. Recall a variable is a container.  When a variable is use in an expression, the value stored in the variable is use to evaluate the expression.

````python
item_cost = 10
tax = 1

total_price = item_cost + tax
print(total_price)
````

Lets sum two number input form the user

````python
item_cost = input("What is it cost of the item?")
tax = input("How much is the tax?")
total_price = item_cost + tax
print(total_price)
````

What happened?  It didn't add the input, it joined(concatenated) the input.  Lets investigate.

````python
item_cost = input("What is it cost of the item?")
print(type(item_cost))
tax = input("How much is the tax?")
print(type(tax))
total_price = item_cost + tax
print(total_price)
````

The data type of the input was string, so the variables were string.  So the *input* function must return a *str* data type.

There is a way we can convert, we can use the *int()* function.    To convert the string we use the *int()* function

> Note: We will look a input validation later in the semester, for now just understand some concepts relate to using different data types

````python
item_cost = input("What is it cost of the item?")
item_cost_int = int(item_cost)
print(type(item_cost_int))
print(item_cost_int)
````

We delve in to function in another notebook, but we can use the output of one function as into to another functions.

````python
item_cost = int(input("What is it cost of the item?"))
print(type(item_cost))
print(item_cost)
````

So our final version of calculating total price is below.  Not this follow a common pattern used in programs
- input
- process
- output

````python
# Input data
item_cost = int(input("What is it cost of the item?"))
tax = int(input("How much is the tax?"))

# Process data
total_price = item_cost + tax

# Output result
print(total_price)
````
