When starting your program,  you may not always know a user's name. So what 
about something like:
* If we know the person's name, display it.
* Else keep saying hello to the entire world.

This is what conditions are all about.

Do you remember print and len functions? Here's another one: input. This one is 
asking the user of your notebook (you) to enter a character 
(which can be a sentence) and store it in a variable.

Here are the steps (the algorithm) to see if we know our user's name:

* Ask our user's name and store his answer in a variable  name .
* Check if the name variable contains a value 
* If it does, say hello to our user using his name. 
* Else, keep saying hello to the world!


How do we check if the name variable contains a values?  We going to use the 
len() function is a non-obvious way.  Recall the len() function returns the 
number if items in the sequence (list, dictionary, string etc.).  The thinking 
is, if the user enters a name, then there will be at least one character or the 
length will be greater than zero. 

```{information}
This is common in programming to use a function or concept is a non obvious way
by interpreting the return value to have a meaning other than the obvious purpose.
```

Here is the actual code:

````python
name = input("What is your name ?")

if len(name) == 0:
    print("Hello World !")
else:
    print("Hello", name)
````

## Boolean data type

In Python, to validate a condition, you use a special data type called boolean. A variable of the boolean type can hold only two values: True  or False.  A boolean is the simplest data type. In Python, boolean types use true/false pairs. Let's declare a couple of boolean variables:


````python
the_course_is_excellent = True
print(the_course_is_excellent)

the_author_is_humble = False
print(the_author_is_humble)
````

What really matters is that the condition expressed after the if keyword must resolve to a boolean. That condition can be expressed as:

1. A True or False value. For instance, if True:.
2. A variable of a boolean data type. For instance, if myVar: where myVar is aboolean type.
3. An expression that resolves to a boolean value, like in the above example.

```{information}
An expressions is anything that produces a value
```

To produce a boolean, you use the [comparison operators](comparison_operators.ipynb).  
Comparison operators are used to compare two values.  Sometimes, you may want 
to have more complicated conditions, where the decision depends on the result 
of a combination of different expressions. This is where you will use 
[logical operators](logical_operators.ipynb). Logical operators allow you 
to combine boolean values: either specific boolean values or results of expressions.

## Chaining Conditions

We've personalized the welcome message, but what if we want to use more that 
one first name? One possibility is to create a chain of conditions. Here is 
the general form: Chaining conditions

```python
if condition1: 
    # instructions
elif condition2:
    # instructions
else:
    # instructions
```


In our "Hello, World" example, the user could send two strings when running the program. We could then send a string containing the concatenation of those two strings:

````python
firstName = input("What is your first name ?")
lastName = input("What is your last name ?")

if len(lastName) > 0 and len(firstName) > 0:
    print("Hello", firstName, lastName)
elif len(firstName) > 0:
    print("Hello", firstName)
else:
    print("Hello World !")
````
