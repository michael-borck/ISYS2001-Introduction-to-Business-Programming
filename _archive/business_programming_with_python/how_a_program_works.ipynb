# How a Program Works

The CPU is an electronic device that is designed to do specific things. In particular, the CPU is designed to perform operations.  See the six operations notebook for a list of simple things a computer can do.

The CPU performs simple operations on pieces of data.  A program is a list of instructions that cause the CPU to perform one of the simple operations.  The instructions are a sequence of 0s and 1s.  These sequences are called machine language.  

Programs are usually stored on secondary devices.  To execute the program it has to be copied into main memory each time a CPU executes it.  This is know as the fetch-decode-execute cycle.
