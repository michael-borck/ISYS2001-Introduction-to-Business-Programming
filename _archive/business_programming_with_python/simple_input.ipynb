# Getting data

Program need data to work on.  Being able to input data interactively at run 
time makes the program more useful.

```{information}
This is an example of the concept of generalisation.  A more generalise program 
can be used in a variety of situations with different input.
```

In Python3 there is a function, input() that allows you to receive input form the keyboard. 

````python
input("What is your name?")
````

That is okay, but we should save the input to a variable so we can use it later.

````python
my_name = input("What is your name?")
print(my_name)
````

Lets use the help() finction to find out more.

````python
help(input)
````

What is the data type of the return value of input()?

````python
my_name = input("What is your name?")
print(my_name)
print(type(my_name))
````

What about inputting numbers?

````python
my_age = input("What is your age?")
print(my_age)
print(type(my_age))
````

Recall from [data type](06_data_types.ipynb) we could convert datatypes.  So lets use the int() function to convert the string into an integer.

Note: We will investigate in later sections how to [validate input](17_input_validation.ipynb), but for now make sure you only enter digits.


````python
my_age = input("What is your age?")
my_age_as_int = int(my_age)
print(my_age_as_int)
print(type(my_age_as_int))
````
