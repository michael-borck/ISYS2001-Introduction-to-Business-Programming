# Programming Languages


## Low Level Languages

Machine language, a sequence of 0s and 1s is very difficult for a person to read and write.  To make it (slightly) easier, assembly language was created.  This use three letter abbreviations.  Assembly language still requires you to know a lot about the CPU.  Any language that requires you to know a lot about the CPU is called a low level language.


## High Level Languages

High level languages allow you to create program without knowing how the CPU works.  Most high level languages use word that are easy to understand.

Each language has it own set of predefined words that you must use to write a program.  These are call keyword or reserved words.  We will discuss Pytohn more in later sections, but here are the list of Pytohn keywords:

|          |         |          |        |
| -------- | ------- | -------- | ------ |
| False    | def     | if       | raise  |
| None     | del     | import   | return |
| True     | elif    | in       | try    |
| and      | else    | is       | while  |
| as       | except  | lambda   | with   |
| assert   | finally | nonlocal | yield  |
| break    | for     | not      |        |
| class    | from    | or       |        |
| continue | global  | pass     |        |



In addition to keywords there are operators, like additions. To combine keywords, operators, data you have to follow a set of rules.  We call these rules syntax.  We will be learning the Python syntax in this course. The individual instructions you write in a high level language are called statements.

Because the CPU only understands machine language. A high level language must be translated to machine language. Depending on the high level language the programmer will use either a complier or and interpretor.

## Complied

A complier is a program that translate a high level language program in a seperate machine language program.  the machine language program can be executes at any time.

## Interpreted

An interpreter is a program the both translate and execute the instructions ina high level language.  A interpreter reads each individual instruction and them immediately translate and executes.  This process is repeated for each instruction in the high level language.  

## Python

Python is an interpreted language, which means the source code of a Python 
program is converted into bytecode that is then executed by the Python virtual 
machine. Python is different from major compiled languages, such as C and C + +, 
as Python code is not required to be built and linked like code for these 
languages. This distinction makes for two important points:

*  Python code is fast to develop
*  Python code is not as fast in execution
