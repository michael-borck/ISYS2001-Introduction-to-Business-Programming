### Hardware

A computer is a collection of interconnected components. A computer's hardware includes all of its physical components. You've probably seen commercials for computer parts like microprocessors, memory, and disc drives. Components are the parts. These areas require extensive computing knowledge to comprehend. It's difficult to understand what these sections do unless you know a lot about computers. a typical computer system consists of the following major components:
* The central processing unit (CPU) - this is the part that actually runs the program.
* Main memory - this is where the computer stores the running program and the data.  This is commonly know a random-access memory or RAM.
* Secondary storage devices - Special type of memory that can store programs and data even when there is no power to the computer.
* Input devices - collects data and sends it to the computer.  Example keyboard, mouse, camera etc
* Output devices - data is sent ot an output device.  Example are monitors or printers.

### Software

Programs are commonly referred to as software. Software is essential to a computer because without software, a computer can do nothing.  There are two general categories of software: 
* System software. Manage the basic operations of the computer hardware
  * Operating systems - fundamental set of programs to control the internal operations ofthe computers hardware and manages all connected devices.
  * Utility Programs - program to perform a specialised task.  Example virus scanners, file compressions
  * Software Development - tools used by the programmer to create, modify and test software.
* Application software - these are program that make the computer useful for everyday taks.  Examples include spreadsheet programs, email programs, web browsers etc.

## How computers store data

A computers memory is divided into tiny storage locations know as bytes.  each byte is divided into eight smaller storage locations known a bits. The bit can indicate one of two values depending on whether it is turned on or off. A bit that is turned off in a computer system represents the number 0 while a bit that is turned on represents the number 1.

### Storing Numbers

A bit can only be used to represent numbers in a very limited fashion. the binary numbering system. In the binary numbering system (or binary, as it is usually called), all numeric values are written as sequences of 0s and 1s. Here is an example of the number 157 that is written in binary:

      100111101

Computers are able to store negative numbers and real numbers in memory, but to do so they use encoding schemes along with the binary numbering system. You don’t need to know how these encoding schemes work, only that they are used to convert negative numbers and real numbers to binary format.


### Storing Characters

Any piece of data that is stored in a computer's memory must be stored as a binary digit. When a character is to be stored into memory it need to be converted into a binary digit.  You don’t need to know how these encoding schemes work, only that they internation standards used to convert binary digits to and from characters.
