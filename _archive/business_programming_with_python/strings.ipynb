## Escape sequence

In Python strings, the backslash " " is a special character, also called the "escape" character. It is used in representing certain whitespace characters: 
- "\t" is a tab, 
- "\n" is a new line, and "\r" is a carriage return. 
- "\\ \\" is the literal backslash character. 

```python
# This is an escape sequence.
string = "This is a \"Google Colab\" python notebook"
print(string)
```

## Accessing a character in the String by an Index

To access a single character in a String, we use the index value.   The index is order the characters are in the sting.  To access a character all you have to do is just place the number (index value) inside the pair of square brackets along with the name of the string.

```python
string = "Python"
print(string[2])
```

```python
print(string[5])
```

## Slicing a string

sometime you want to access a set f characters within a string.  Slicing a string helps to get a set of characters, a sub-string, from a string.  Below are some slicing examples that are useful.

```python
string = "programming"
string
```

**Getting  one character of the string**

```python
print(string[0:1])
```

**Getting the first three characters from the string**

```python
print(string[0:3])
```

**Getting the first three characters from the string (Alternate)**

```python
print(string[:3])
```

**Getting the last three characters from the string**

```python
print(string[-3:])
```

**Getting all characters but excluding three first characters from the string**

```python
print(string[3:])
```

**Getting all characters but excluding the last three characters from the string**

```python
print(string[:-3])
```

**Reversing all the characters in a given string**

```python
print(string[::-1])
```

**Alternative to printing all characters in a string**

```python
print(string[::])
```

## Replacing a string

The Replace function in python is one of the best function that can be applied to strings. For example, shown below is a string "Money" we need to replace the dollar and the comma sign and this can be done as shown below.

```python
Money = '$113,678'
print(Money)
print("===========================")
print(type(Money))
```

```python
Money = Money.replace('$', '')
Money
```

```python
Money = Money.replace(',', '')
Money
```

## Converting strings to integers

There are other data types in Python like integers.  If you know a string only contains numbers, then can convert form a string to a numerical data type.  In the following example, we convert the money data type to an integer.

*Note: Taking a sting and removing unwanted character and then converting to a number to use later in your program is a common strategy when programming.  We will discuss strategies and lagorithms later in the semester*

```python
Money = int(Money)
print(Money)
print("===========================")
print(type(Money))
```

## The Join method

The join function is python is used to join the string according to the specified pattern.

```python
String = "Python Programming is fun"
String
```

## Capitalize

The capitalize function capitalizes the first character in the word or a string.

```python
string = "programming"
string
```

```python
string = string.capitalize()
string
```

## The Find method

The find method returns the index of the given substring position. If the value is not found it returns, -1.

```python
string = "programming"
string
```

```python
print(string.find('p'))
```

```python
print(string.find('t'))
```

## The Strip method

The strip function strips or removes the white spaces both from the starting and the ending of a string.

```python
string = "     programming is easy    "
string
```

```python
print(string.strip())
```
