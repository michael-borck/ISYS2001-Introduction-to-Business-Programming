# Case Study:  Count the Vowels

We looked at variables and their values. In Python we you don't need to define 
the  variable's type as it depends on the value stored within. Storing money 
requires a different container from the one needed for a book. In Python, 
the storage space of a container automatically adapts itself to content.

Lets write an application and need to analyze some text and calculate the following:
- The number of vowels present in the text. 
- The percentage of vowels. 

You could break down the process into the following:
- Ask user for some text.
- Browse the text supplied by the user; character by character.
- Increase your total every time you find a vowel.
- Divide the final number of vowels by the total number of characters in the string to get the percentage.
- Multiply the result by 100 for the final percentage.

Imagine having to do that by hand with this document! How many variables do 
you need to implement this? Think about how many individual pieces of 
information you need to store:

- The initial sequence of characters you ask the user for (a string).
- The total of the number of vowels in the string.
- The percentage of vowels.

It looks like you need three variables! To define each of them, you need the following components:

- A name, so you can find the information later.
- An initial value, which will give you a starting point, and the type!

See [data type](data_types.ipynb) section for information of the different type. For this problem we use:
- String = text going to analyse
- Integer = whole number to count the number of vowels
- Float = floating point number for the percentage

```python
# Initialise variable
text = "Hello World!"
number_of_vowels = 0
percentage_of_vowels = 0.0

# Look at each letter in our text, if it is a vowel, increment our counter
for letter in text:
  if letter.capitalize() in ['A','E','I','O','U']:
    number_of_vowels += 1

# Calculate the percentage
percentage_of_vowels = ( number_of_vowels / len(text) ) * 100

# Output the result
print("The number of vowels found: ", number_of_vowels)
print("The percentage of vowels calculate: ", percentage_of_vowels, '%')
```
