# Modules

Let's assume that you need to calculate the square root of a number for one of 
your programs. There is no existing square root function in Python. You could 
write one, but there have been a lot of people before you in the exact same 
position. One of them already created a function.  So everyone can use the 
function it is packaged into a module!

## What is a module

A module is like a code library: a file containing a set of functions, classes, 
and variables you want to include in your application.

You use the *import* keyword to import the module.

## What is a package

A package is a collection of Python modules. While a module is a single Python 
file, a package is a directory containing an additional __init__.py file. This 
distinguishes it from a directory that contains a number of scripts.

You have to use the *'.'* operator to access a module after importing the whole 
package

## NumPy: our first package.

Let's try this with numpy, a well-known package containing a lot of numerical 
tools! To import the NumPy package, you could write ```import numpy```, 
but it's convention

    import numpy as np

What about the square root function mentioned above? It's the ```sqrt()```  function of NumPy:

    np.sqrt(16) # -> 4.0


````python
import numpy as np
result = np.sqrt(16)
print(result)
````

````python
import numpy as np
number = int(input("Please input a whole number "))
result = np.sqrt(number)
print("The square root of " + str(number) + " is " + str(result))
````
