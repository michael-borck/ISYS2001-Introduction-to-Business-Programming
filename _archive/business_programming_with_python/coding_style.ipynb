# Coding Style

There are many ways to write equally valid code and, over the years, software developers have worked out ways that help to make writing, using, debugging, and reading code more fun and certainly more efficient. **Following these best practice tips will save you hours and *hours* of time.**

Most of the time, you are going to be writing code that someone else will read (most probably it will be future you), or that you'll need to run again. If you follow the guidelines in this chapter you will find that your code will be:

 - easier to understand
 - more efficient
 - easier to maintain, scale, debug, and refactor


## Code in Style

Python has a whole set of conventions about good style called [PEP8](https://www.python.org/dev/peps/pep-0008/). It includes advice like indentation should always be 4 spaces (not tabs) per level and that you should surround top-level function and class definitions with two blank lines.

Learning *all* of the PEP8 conventions would be tedious beyond belief. There are packages that can check, re-style and even re-write your code automatically.

See the associated sections of the important conventions we are following in this unit:
* Naming Conventions
* Whitespace
* Line width and continuations


## Clean Code

While automation can help apply style, it can't help you write *clean code*. 
Clean code is a set of rules and principles that helps to keep your code 
readable, maintainable, and extendable.  See the notebook on [Principle of Clean Code](principles_of_clean_code.ipynb) for more details.

## Code Comments

Code comments, extra information that is not executed when the code is run, 
can be added by a preceding hash character `# This is a comment`. Use code 
comments to provide extra contextual information that *isn't* conveyed by 
function and variable names.  See [Comments](comments.ipynb) for more details.
