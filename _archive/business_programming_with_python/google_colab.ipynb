# Programming Environment

Your *programming environment* is the computer and all the software that's installed on your system which helps you write and run programs.  In this unit we use a cloud based programming environment.  This section will help you get your system set up so you can start writing programs quickly.

## Overview

Our goal of this notebook is to help you get Python up and running on your computer, so that you can write and run your own programs. To do this, we want to:

- Install Jupyter Notebooks
  - *Google Colab*, zero install, browser based IDE
  - Anaconda, an extensive Python environment install locally
- Help you enter and run Hello World, your first Python program.
- Congratulate you if it works, and give you some options for troubleshooting if it doesn't work.

## Google Colab


Colaboratory, or “Colab” for short, is a product from Google Research. Colab allows anybody to write and execute arbitrary python code through the browser based integrated development environment (IDE).  This is our preferred option in this course.

You will need a free google account to be able to store you notebook in Google Drive, or can be loaded for [GitHub](https://github.com/).  See [Colaboratory Frequently Asked Questions](https://research.google.com/colaboratory/faq.html)

Open a browser of your choice and go to [colab.research.google.com](https:colab.research.google.com) and sign in using your Google account. Click on a new notebook to create a new runtime instance.

## Installing Python on your own computer¶


If you want to install Python on your own computer, we highly recommend you install Python through the Anaconda distribution. In the box below, you can find detailed installation instructions (thanks to the Netherlands eScience Center) specific to your operating system (Mac, Windows, or Linux).

Anaconda as it will install Python, Jupyter Notebook, and other commonly used scientific computing packages and data projects onto your computer.  To begin, download the [Anaconda distribution](https://www.anaconda.com/distribution/)  Then double-click the downloaded file to install. Once installed, Launch Anaconda Navigator
- On Mac, launch through Launchpad or find in application folder
- On Linux, form command line type '''anaconda-navigator'''
- On Windows, launch form your program list 

Once this is done, you will have several applications panels.  The next step is to launch Jupyter Notebook.
