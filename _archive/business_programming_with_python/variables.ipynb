# What is a variable?


It help to read the notebook on Values.  Computers do two things really well. 
***Store values*** and ***perform operations*** on those values. Programs do 
something with data.  Data is stuff that you put into your program.  
We use variables to store, manipulate and process the data.  A variable is a 
container.  You can think of a variable as a box that contains a value.  
This box is stored on a shelf within a gigantic warehouse.  In this analogy, 
the warehouse is the computers memory (or RAM).  

A **value** is a piece od data we put into a variable.  Keeping with our warehouse analogy, different boxes and store different thing.  Similarly, different variable can store different values. As in the ware house you can change the content of boxes we can do similar things to variables.

The location of every box is carefully recorded.  In a computer this location is a memory address.  Memory address are hard for human to remember and use, so we give our variable a unique name.  We call this name an identifier.

## Naming variables

#### Naming rules

Variable names must be unique and they are case sensitive.
* Only letters, numbers, and the underscore character (_) can be used  ```this.is.not.allowed```
* No spaces ```this is not allowed```
* No quotes ```“NotAllowed”```
* Can’t start with a number ```6notAllowed```
* Can't be a keyword  ```import```

#### Naming conventions
We will learn other naming conventions, but for now
* variable names should be meaningful (can someone else understand)
* separate words with underscores ```monthly_sales```  (this is called snake_case)

## Creating a variable

You must create a variable in code if you are going to use one. Python is a  object-oriented language, meaning that every variable is an object.  We discuss object oriented programming later in the semester. 

Variables need a name and contain a value.  Rather than declaring the type of value a variable can store (need to know how much memory space to reserve), in Python they are declared automatically when you assign a value to a variable.

There are several types of variables in Python. Variables containing numbers can be declared by writing the name of a variable and its initial value.

````python
temperature = 27.5
print(temperature)
print(type(temperature))
````

For example in the above cell we we declared a variables _temperature_ with a value _27.5_.  We used the print() function and the type() function to output the value and the datatype 'float'.  See the notebooks on values and datatype for more information.


## Changing the value of a variable

A variable may vary, that is **change is value**.  You can do this by performing a number of **operations**.  See the operations notebook for more examples.


````python
# initialise our savings
savings = 10

# add 100 to our savings
saving = savings + 100

# Calculate the number of days to save to our goal
number_of_days_to_save = (5000 - savings) / 100 #assume saving $100 per day

print(number_of_days_to_save)
````

That is our largest block of code to date. Lets review some of the conepts form this and other notebooks.  You'll notice that the code doesn't all look the same. That's because it's composed of comments and statements:
- Lines that start with # are comments. They are used to help another person understand the code.
- Lines of code that do things are called statements. As the name implies, each statement tells the computer to do something.

Here, each of the statements assigns a value to a variable. The assignment operator is ```=```.

- On the right of the assignment operator, you write an expression which is a statement that produces a value.

- On the left of the assignment operator, you write the name of the variable you want the result to correspond to.

To review, to assign a value to a variable, you write a statement. This statement consists of the name of the variable, followed by the assignment operator, and finally, the expression that produces a value that matches the type of the variable.
