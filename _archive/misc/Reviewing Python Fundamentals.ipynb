## Dictionaries and Sets

```python
# Dictionaries and Sets
```

```python
# Dictionaries are a fundamental data structure in Python that store key-value pairs.
# They are mutable, unordered, and can be nested.
```

```python
# Creating a dictionary
my_dict = {'name': 'John', 'age': 30, 'city': 'New York'}
```

```python
# Accessing dictionary values
print(my_dict['name'])  # Output: John
```

```python
# Modifying dictionary values
my_dict['age'] = 35
print(my_dict['age'])  # Output: 35
```

```python
# Adding new key-value pairs
my_dict['occupation'] = 'Engineer'
print(my_dict)  # Output: {'name': 'John', 'age': 35, 'city': 'New York', 'occupation': 'Engineer'}
```

```python
# Removing key-value pairs
del my_dict['city']
print(my_dict)  # Output: {'name': 'John', 'age': 35, 'occupation': 'Engineer'}
```

```python
# Checking if a key exists
print('name' in my_dict)  # Output: True
```

```python
# Iterating over a dictionary
for key, value in my_dict.items():
    print(key, value)
```

```python
# Sets are another data structure in Python that store unique elements.
# They are mutable, unordered, and do not allow duplicate values.
```

```python
# Creating a set
my_set = {1, 2, 3, 4, 5}
```

```python
# Adding elements to a set
my_set.add(6)
print(my_set)  # Output: {1, 2, 3, 4, 5, 6}
```

```python
# Removing elements from a set
my_set.remove(3)
print(my_set)  # Output: {1, 2, 4, 5, 6}
```

```python
# Checking if an element exists
print(2 in my_set)  # Output: True
```

```python
# Set operations
set1 = {1, 2, 3}
set2 = {3, 4, 5}
```

```python
# Union
union_set = set1.union(set2)
print(union_set)  # Output: {1, 2, 3, 4, 5}
```

```python
# Intersection
intersection_set = set1.intersection(set2)
print(intersection_set)  # Output: {3}
```

```python
# Difference
difference_set = set1.difference(set2)
print(difference_set)  # Output: {1, 2}
```

## File Handling

```python
# File Handling
```

```python
# Reading from a file
file_path = "example.txt"  # Specify the path of the file
```

```python
try:
    with open(file_path, "r") as file:  # Open the file in read mode
        content = file.read()  # Read the entire content of the file
        print(content)  # Print the content
except FileNotFoundError:
    print("File not found!")  # Handle file not found error
```

```python
# Writing to a file
file_path = "output.txt"  # Specify the path of the file
```

```python
try:
    with open(file_path, "w") as file:  # Open the file in write mode
        file.write("Hello, World!")  # Write content to the file
    print("File written successfully!")
except:
    print("Error occurred while writing to the file!")  # Handle any errors
```

## Exception Handling

```python
try:
    num1 = 10
    num2 = 0
    result = num1 / num2
    print(result)
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
```

```python
try:
    num1 = 10
    num2 = "0"
    result = num1 / num2
    print(result)
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
except TypeError:
    print("Error: Invalid operand types.")
```

```python
try:
    num1 = 10
    num2 = "0"
    result = num1 / num2
    print(result)
except (ZeroDivisionError, TypeError):
    print("Error: Invalid input.")
```

```python
try:
    num1 = 10
    num2 = 2
    result = num1 / num2
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
else:
    print("The result is:", result)
```

```python
try:
    num1 = 10
    num2 = 0
    result = num1 / num2
except ZeroDivisionError:
    print("Error: Cannot divide by zero.")
finally:
    print("Done.")
```
